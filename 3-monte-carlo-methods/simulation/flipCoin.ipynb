```python
# import required libraries

import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coin_flip():
    return random.randint(0, 1)
```

```python
# check the return value of coin_flip
coin_flip()
```

```python
# Monte Carlo simulation;

# empty list to store the probability values.
probab_val = []

def monte_carlo(n):
    results = 0
    for i in range(n):
        flip_result = coin_flip()
        results += flip_result

        # calculatig probability value and appending it to the list.
        probab_val.append(results/(i+1))

        # plotting the graph
        plt.axhline(y=0.5, color='r', linestyle='-')
        plt.xlabel("Number of flips")
        plt.ylabel("Probability of heads")
        plt.plot(probab_val)
    return results/n
```

```python
# check the return value of monte_carlo
answer = monte_carlo(500)

print("Final probability value: ", answer)
```
